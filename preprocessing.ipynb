```python
from bs4 import BeautifulSoup
import re
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

def prepros_word(raw_text):
    
    print("----------------------Kalimat Asli----------------------")
    print(raw_text)
    
    #1. Case folding
    lowertext = raw_text.lower()
    print("\n----------------------Case Folding----------------------")
    print(lowertext)
    
    #2. Filtering
    deleteurl = re.sub(r"http\S+", "", lowertext)
    letters_only = re.sub("[^a-zA-Z]", " ",str(deleteurl))
#     spacetext = letters_only.strip()
    spacetext = lowertext.strip()
    print("\n----------------------Filtering----------------------")
    print(spacetext)
 
    #3. Tokenisasi
    words = spacetext.split()
#    words1 = word_tokenize(spacetext)
    print("\n----------------------Tokenisasi----------------------")
    print(words)
    
    
    #4. Stemming
    stem_words=[]
    stem1 = StemmerFactory()
    stemmer = stem1.create_stemmer()
    for word in words:
        stem_text = stemmer.stem(word)
        stem_words.append(stem_text)
    print("\n----------------------Stemming----------------------")
    print(stem_words)
        
    #5. Remove stop words
    stopword_factory = StopWordRemoverFactory()
    stopwords = stopword_factory.get_stop_words()
    filtered_text = [word for word in stem_words if word not in stopwords]
    print("\n----------------------Remove Stopword----------------------")
    print(filtered_text)
    
    print("\n----------------------Hasil Akhir Setelah PreProcessing----------------------")
    # 5. Join the words back into one string separated by space & return, remove the short text
    return(" ".join(w for w in filtered_text if len(w)>2))
```

```python
stopword_factory = StopWordRemoverFactory()
stopwords = stopword_factory.get_stop_words()
stopwords
```

```python
df_ir["clean_text"] = df_ir["Title"].apply(prepros_word) + " " + df_ir["Desc"].apply(prepros_word)
df_ir
```

```python
def get_want_to(text):
#     print(text)
    search_words = ['tawarkan', 'tawakno', 'cari', 'wtb','dapat','oleh','yg jual', 'yang jual']
    if any(word in text for word in search_words):
        wt = "wtb"
    else: 
        wt = "wts" 
    return wt
    
# get_want_to(clean_text)
df_ir["wt"] = df_ir["clean_text"].apply(get_want_to)
df_ir
```

```python
def get_cat(text):
#     print(text)
    if_laptop = ['windows','core','laptop','leptop','notebook','netbook','asus','acer','toshiba','axioo','dell','lenovo']
    if any(word in text for word in if_laptop):
        cat = "laptop"
    else: 
        cat = "" 
    return cat
    
# get_want_to(clean_text)
df_ir["category"] = df_ir["clean_text"].apply(get_cat)
df_ir
```
